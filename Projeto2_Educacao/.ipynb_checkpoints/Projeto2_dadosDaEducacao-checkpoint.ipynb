```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("StudentsPerformance+(1).csv")
```

## Percentages

```python
#porcetagem de mulheres e homens
gender_percent = df['gender'].value_counts(normalize=True)*100

plt.bar(gender_percent.index,gender_percent)
plt.ylabel("percent");
```

```python
#porcetagem de etinia
race_percent = df['race/ethnicity'].value_counts(normalize=True)*100
```

```python
plt.bar(race_percent.index,race_percent)
plt.ylabel("percent");
```

```python
#porcetagem parental level of education
par_level_percent = df['parental level of education'].value_counts(normalize=True)*100
```

```python
plt.bar(par_level_percent.index,par_level_percent)
plt.xticks(rotation=50,ha='right');
plt.ylabel("percent");
```

```python
#porcetagem lunch
lunch_percent = df['lunch'].value_counts(normalize=True)*100
```

```python
plt.bar(lunch_percent.index,lunch_percent)
plt.ylabel("percent");
```

```python
#porcetagem test preparation
test_prep_percent = df['test preparation course'].value_counts(normalize=True)*100
```

```python
plt.bar(test_prep_percent.index,test_prep_percent)
plt.ylabel("percent");
```

## Bloxplots

```python
fig, axs = plt.subplots(1,3,figsize = (15,4))

sns.boxplot(data=df,x="writing score",y='gender',ax=axs[0])
sns.boxplot(data=df,x="reading score",y='gender',ax=axs[1])
sns.boxplot(data=df,x="math score",y='gender',ax=axs[2])

plt.tight_layout()
```

```python
df.groupby('gender').mean()
```

```python
df[['gender','math score']].groupby('gender').describe()
```

```python
fig, axs = plt.subplots(1,3,figsize = (15,4))

sns.boxplot(data=df,x="writing score",y='race/ethnicity',ax=axs[0])
sns.boxplot(data=df,x="reading score",y='race/ethnicity',ax=axs[1])
sns.boxplot(data=df,x="math score",y='race/ethnicity',ax=axs[2])

plt.tight_layout()
```

```python
fig, axs = plt.subplots(1,3,figsize = (15,4))

sns.boxplot(data=df,x="writing score",y='parental level of education',ax=axs[0])
sns.boxplot(data=df,x="reading score",y='parental level of education',ax=axs[1])
sns.boxplot(data=df,x="math score",y='parental level of education',ax=axs[2])

plt.tight_layout()
```

```python
fig, axs = plt.subplots(1,3,figsize = (15,4))

sns.boxplot(data=df,x="writing score",y='lunch',ax=axs[0])
sns.boxplot(data=df,x="reading score",y='lunch',ax=axs[1])
sns.boxplot(data=df,x="math score",y='lunch',ax=axs[2])

plt.tight_layout()
```

```python
fig, axs = plt.subplots(1,3,figsize = (15,4))

sns.boxplot(data=df,x="writing score",y='test preparation course',ax=axs[0])
sns.boxplot(data=df,x="reading score",y='test preparation course',ax=axs[1])
sns.boxplot(data=df,x="math score",y='test preparation course',ax=axs[2])

plt.tight_layout()
```
